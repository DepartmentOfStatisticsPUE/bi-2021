Instalujemy pakiet `gtrendsR`

```r
install.packages("gtrendsR")
```

Ładujemy pakiet

```r
library(gtrendsR)
```

Podstawowa funkja w pakiecie `gtrendsR` to `gtrends`

```r
wynik <- gtrends(keyword = c("Uniwersytet Ekonomiczny w Poznaniu", 
                             "Uniwersytet Ekonomiczny w Krakowie"),
                geo = "PL",
                time = "today+5-y")

str(wynik,1)
```

```r
head(wynik$interest_over_time)
```

```r
plot(wynik)
```
